```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier, plot_tree
import matplotlib.pyplot as plt

data = {
    'Engine_Size': [1800, 2400, 1500, 3500, 2200, 1300, 2600, 1600, 1800, 2500, 1700, 3000, 1400, 2800, 2000, 3200, 2300, 1200, 2700, 1900],
    'Curb_Weight': [1400, 1800, 1200, 2100, 1600, 1100, 1900, 1300, 1250, 2000, 1500, 2200, 1150, 1950, 1700, 2050, 1550, 1050, 1850, 1650],
    'Engine_Type': ['Gasoline', 'Hybrid', 'Electric', 'Hybrid', 'Gasoline', 'Electric', 'Hybrid', 'Gasoline', 'Gasoline', 'Hybrid', 'Electric', 'Gasoline', 'Gasoline', 'Hybrid', 'Electric', 'Gasoline', 'Hybrid', 'Electric', 'Gasoline', 'Hybrid'],
    'Price_Range': ['Medium', 'High', 'High', 'High', 'High', 'Low', 'High', 'Medium', 'Low', 'High', 'Medium', 'High', 'Low', 'High', 'Medium', 'High', 'High', 'Low', 'High', 'Medium']
}

df = pd.DataFrame(data)

label_encoder = LabelEncoder()
df['Engine_Type'] = label_encoder.fit_transform(df['Engine_Type'])
df['Price_Range'] = label_encoder.fit_transform(df['Price_Range'])

X = df[['Engine_Size', 'Curb_Weight', 'Engine_Type']]
y = df['Price_Range']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state = 4)

model = DecisionTreeClassifier()
model.fit(X_train, y_train)

accuracy = model.score(X_test, y_test)
print(f"Accuracy: {accuracy}")

plt.figure(figsize=(10,10))
plot_tree(model, filled=True, feature_names=['Engine_Size', 'Curb_Weight', 'Engine_Type'], class_names=['Low', 'Medium', 'High'])
plt.show()
```
