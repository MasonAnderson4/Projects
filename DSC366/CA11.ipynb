```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

z = np.random.normal(size=100)
categories = ['Apple', 'Banana', 'Cherry', 'Date']

bar_heights = np.random.randint(1, 15, size=4)
bar_heights2 = np.random.randint(1, 15, size=4)

plt.figure(figsize=(14, 10))

plt.subplot(2, 2, 1)
plt.plot(x, y, label='Sin x')
plt.title('line plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)

plt.subplot(2, 2, 2)
sizes = 50 * abs(y)
plt.scatter(x, y, s=sizes, c=sizes, cmap='viridis', alpha=0.6, edgecolor='black', label='sinx')
plt.title('scatter plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()

plt.subplot(2, 2, 3)
plt.hist(z, bins=20, color='green', label='Normal Distribution')
mean_z = np.mean(z)
plt.axvline(mean_z, color='red', linestyle='dashed', linewidth=1)
plt.text(mean_z, 10, 'Mean: {:.2f}'.format(mean_z), rotation=0, color='red')
plt.title('Histogram')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.legend()

plt.subplot(2, 2, 4)
index = np.arange(len(categories))
bar_width = 0.35
plt.bar(index, bar_heights, bar_width, label='Set 1', color='b')
plt.bar(index + bar_width, bar_heights2, bar_width, label='Set 2', color='r')
plt.title('Bar Chart Comparison')
plt.xlabel('Categories')
plt.ylabel('Counts')
plt.xticks(index + bar_width / 2, categories)
plt.legend()

plt.tight_layout()
plt.show()
```
