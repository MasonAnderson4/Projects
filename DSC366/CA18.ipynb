```python
import matplotlib.pyplot as plt
#Initialize Parameters:
#Set the initial weight to 0.3.
#Define the input score as 2.
#Set the ground truth prediction to 0.4.
#Define the learning rate (alpha) as 0.1.

weight = 0.3
input_score = 2
ground_truth_prediction = 0.4
alpha = 0.1

#2. Gradient Descent Simulation:

errors = []
predictions = []

#Use a simple linear model prediction = input_score * weight for prediction.
#Compute the loss using mean squared error: loss = (prediction - ground_truth) ** 2.
#Update the weight by computing the gradient of the loss with respect to the weight and adjust the weight accordingly.
#Perform this update in a loop for 20 iterations, printing the error and the current prediction after each update.

for x in range(20):
  prediction = input_score * weight
  loss = (prediction - ground_truth_prediction) ** 2

  errors.append(loss)
  predictions.append(prediction)

  direction_and_amount = (prediction - ground_truth_prediction) * input_score
  weight = weight - (alpha * direction_and_amount)
  print("Iteration:", x + 1)

  print("Error:", loss)
  print("Prediction:", prediction)



# 3. Visualizing the Learning Process:

#Plot the loss over iterations using matplotlib to visualize how the loss decreases as the weight is adjusted.

plt.plot(errors, marker = 'o')
plt.title("Loss over Iterations")
plt.xlabel("Iterations")
plt.ylabel("Error")
plt.show()

```
